```julia
using BSON
using JSServe
using WGLMakie
Page(exportable=true, offline=true)
BSON.@load "vae.bson" μ y
BSON.@load "vae.bson" μ3 y3
BSON.@load "e.bson" e
BSON.@load "encoded_outputs.bson" encoded_outputs
BSON.@load "y2.bson" y2
e = cat([Float32.(e[i]') for i∈[1,2,3]]...,dims=1)
v = cat([Float32.(encoded_outputs[i][1][1]) for i∈1:1000]...,dims=2)
y2 = Float32.(y2);
w = cat(v[:,1:997],e,dims=2)
z = Int.(cat(y2[1:997],-2*ones(Float32,27,1),dims=1)[:]);
ms = [0.1*ones(997,1);0.3*ones(27,1)][:];
```

```julia
WGLMakie.activate!()
meshscatter(μ[1,:],μ[2,:],μ[3,:],color=y, colormap=:darktest,markersize=0.2)
```

```julia
WGLMakie.activate!()
meshscatter(w[1,:],w[2,:],w[3,:], color=z, colormap=:darktest,markersize=ms)
```
